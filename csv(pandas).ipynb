```python
#USING CSV PACKAGE
import csv
csv_columns = ['Name','Age','Address']
dict = [
    {'Name':'Aayush','Age':23,'Address': 'KTM'},
    {'Name':'Shyam','Age':45,'Address': 'BKT'}
]
csvfile = open('data.csv','w')
writer = csv.DictWriter(csvfile,fieldnames=csv_columns)
writer.writeheader()
for data in dict:
    writer.writerow(data)
csvfile.close()

```
